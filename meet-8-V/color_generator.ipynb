```python
# Function to convert RGBA to HEX in the format 0xffRRGGBBAA
def rgba_to_hex(rgba_str):
    rgba = rgba_str.strip("rgba() ").split(",")
    r, g, b, a = map(float, rgba)
    # Convert alpha to 255 scale
    alpha = int(a * 255)
    return f"0xff{int(r):02x}{int(g):02x}{int(b):02x}{alpha:02x}"

# Convert the dictionary values to the desired HEX format
hex_dict = {key: rgba_to_hex(value) for key, value in css_dict.items()}

hex_dict
```

```python
def transform_to_camel_case(key):
    parts = key.split("*")[-1].split("-")
    return parts[0] + ''.join(word.capitalize() for word in parts[1:])

complete_transformed_data = {
    (transform_to_camel_case(k) if "*" in k else k): v for k, v in hex_dict.items()
}

complete_transformed_data
```

```python
classes = {}

for key, value in complete_transformed_data.items():
    if any(str(i) in key for i in range(25, 1001, 25)):
        value_cleaned = value[:-2]  # Removing the last two characters (ff)
        class_name = f"_{key.split('Color')[0].capitalize()}Colors"
        if class_name not in classes:
            classes[class_name] = []
        classes[class_name].append(f"  final {key} = const Color({value_cleaned});")

# Generating the classes dynamically
output = ""
for class_name, definitions in classes.items():
    output += f"class {class_name} {{\n"
    output += f"  {class_name}();\n\n"
    output += "\n".join(definitions)
    output += "\n}\n\n"

print(output)
```
