# Challenge: Program Loop FizzBuzz Berdasarkan 3 Digit Terakhir NPM

## 1. Deskripsi Tantangan

Buat program yang menggunakan **loop** untuk menampilkan angka dari 1 sampai dengan **3 digit terakhir NPM**. Program ini akan melakukan pengecekan terhadap setiap angka dan menampilkan hasil yang berbeda berdasarkan kondisi tertentu.

### Kondisi Pengecekan:
1. Jika angka merupakan **kelipatan 3**, cetak "Fizz".
2. Jika angka merupakan **kelipatan 5**, cetak "Buzz".
3. Jika angka merupakan **kelipatan 3 dan 5 sekaligus**, cetak "FizzBuzz".
4. Jika tidak memenuhi salah satu kondisi di atas, cetak angka tersebut.

## Penyelesaian:

```python
# Langkah 1: Tentukan 3 digit terakhir NPM (contoh NPM: 5210411135 -> 135)
npm_terakhir = 135

# Langkah 2: Loop dari 1 hingga nilai 3 digit terakhir NPM
for angka in range(1, npm_terakhir + 1):
    if angka % 3 == 0 and angka % 5 == 0:
        print("FizzBuzz")
    elif angka % 3 == 0:
        print("Fizz")
    elif angka % 5 == 0:
        print("Buzz")
    else:
        print(angka)
```
